```python
import os
import shutil
import random
import string

thu_muc_nguon = r"D:\ML2\anhgoc\vit_val"
thu_muc_dich = r"D:\ML2\dataset\images\val"

os.makedirs(thu_muc_dich, exist_ok=True)

i = 1
for ten_file in os.listdir(thu_muc_nguon):
    duong_dan_cu = os.path.join(thu_muc_nguon, ten_file)

    if ten_file.lower().endswith((".jpg", ".jpeg", ".png", ".webp")):

        # tạo chuỗi random gồm chữ + số
        chuoi_random = ''.join(random.choices(
            string.ascii_letters + string.digits, k=6
        ))

        # tên mới: bắt đầu bằng số thứ tự + chữ + số random
        ten_moi = f"{i}_{chuoi_random}.jpg"

        duong_dan_moi = os.path.join(thu_muc_dich, ten_moi)

        shutil.copy(duong_dan_cu, duong_dan_moi)

        i += 1
```

```python
from ultralytics import YOLO
import cv2
model = YOLO(r"D:\ML2\yolo_results\weights\best.pt")
cap = cv2.VideoCapture(r"D:\ML2\demvo.mp4")
line1 = (0, 500)
line2 = (720, 500)
counted_ids = set()
duck = 0
chicken=0
def get_side(px, py, x1, y1, x2, y2):
    return (px-x1)*(y2-y1) - (py-y1)*(x2-x1)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    h, w, c = frame.shape
    # frame = cv2.resize(frame, (720, 1080))
    cv2.line(frame, line1, line2, (0, 255, 0), 1)
    results = model.track(
        frame,
        tracker=r"D:\ML2\custom_bytetrack.yaml",
        conf=0.2,
        persist=True,verbose=False
    )

    boxes = results[0].boxes

    if boxes.id is not None:

        for box, track_id in zip(boxes, boxes.id):

            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            track_id = int(track_id)

            class_name = model.names[cls_id]

            
            cx = int((x1 + x2) / 2)
            cy = int((y1 + y2) / 2)

        
            cv2.circle(frame, (cx, cy), 5, (0,255,255), -1)

            
            now_side = get_side(cx, cy, line1[0], line1[1], line2[0], line2[1])

            if now_side<0 and (track_id not in counted_ids):
                if class_name == "chicken":
                    chicken+=1
                elif class_name == "duck":
                    duck+=1
                counted_ids.add(track_id)

            # chọn màu
            if class_name == "chicken":
                color = (0, 255, 0)
            elif class_name == "duck":
                color = (0, 0, 255)
            if track_id  in counted_ids:
                color = (19, 122, 255)

            # vẽ box
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            label = f"Id {track_id} {class_name}  {conf:.2f}"

            cv2.putText(
                frame,
                label,
                (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.6,
                color,
                2
            )

    # hiển thị count
    cv2.putText(
        frame,
        f"duck: {duck}",
        (30, 50),
        cv2.FONT_HERSHEY_SIMPLEX,
        1.2,
        (0, 0, 255),
        3
    )
    cv2.putText(
        frame,
        f"chicken: {chicken}",
        (30, 80),
        cv2.FONT_HERSHEY_SIMPLEX,
        1.2,
        (0, 255, 0),
        3
    )

    cv2.imshow("YOLO Tracking Chicken & Duck", frame)

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
model = YOLO(r"D:\ML2\yolo_results\weights\best.pt")
cap = cv2.VideoCapture(r"D:\ML2\demvo.mp4")
roi_x1 = 250
roi_y1 = 300
roi_x2 = 720
roi_y2 = 600
counted_ids = set()
previous_state = {}
duck = 0
chicken=0
def is_inside_roi(cx, cy):
    return roi_x1 < cx < roi_x2 and roi_y1 < cy < roi_y2
while True:
    ret, frame = cap.read()
    if not ret:
        break
    h, w, c = frame.shape
    frame = cv2.resize(frame, (720, 1080))
    cv2.rectangle(frame,
              (roi_x1, roi_y1),
              (roi_x2, roi_y2),
              (255, 255, 0),  # màu vàng nhạt (BGR)
              1) 
    results = model.track(
        frame,
        tracker=r"D:\ML2\custom_bytetrack.yaml",
        conf=0.2,
        persist=True,verbose=False
    )

    boxes = results[0].boxes

    if boxes.id is not None:

        for box, track_id in zip(boxes, boxes.id):

            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            track_id = int(track_id)

            class_name = model.names[cls_id]

            
            cx = int((x1 + x2) / 2)
            cy = int((y1 + y2) / 2)

        
            cv2.circle(frame, (cx, cy), 5, (0,255,255), -1)

            
            # now_side = get_side(cx, cy, line1[0], line1[1], line2[0], line2[1])

            inside_now = is_inside_roi(cx, cy)
            inside_prev = previous_state.get(track_id, False)

            if not inside_prev and inside_now and track_id not in counted_ids:
                if class_name == "chicken":
                    chicken += 1
                elif class_name == "duck":
                    duck += 1

                counted_ids.add(track_id)

            previous_state[track_id] = inside_now

            # chọn màu
            if class_name == "chicken":
                color = (0, 255, 0)
            elif class_name == "duck":
                color = (0, 0, 255)
            if track_id  in counted_ids:
                color = (19, 122, 255)

            # vẽ box
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            label = f"Id {track_id} {class_name}  {conf:.2f}"

            cv2.putText(
                frame,
                label,
                (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.6,
                color,
                2
            )

    # hiển thị count
    cv2.putText(
        frame,
        f"duck: {duck}",
        (30, 50),
        cv2.FONT_HERSHEY_SIMPLEX,
        1.2,
        (0, 0, 255),
        3
    )
    cv2.putText(
        frame,
        f"chicken: {chicken}",
        (30, 80),
        cv2.FONT_HERSHEY_SIMPLEX,
        1.2,
        (0, 255, 0),
        3
    )

    cv2.imshow("YOLO Tracking Chicken & Duck", frame)

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
