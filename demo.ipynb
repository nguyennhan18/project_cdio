# 🎥 Poultry Detection Demo
File này dùng để chạy thử nghiệm nhanh mô hình nhận diện Gà & Vịt.

```python
from ultralytics import YOLO
from IPython.display import display, Image

# 1. Load mô hình đã huấn luyện
model = YOLO('weights/best.pt')

# 2. Đường dẫn file cần test (Thay đổi tên file của bạn tại đây)
SOURCE_PATH = 'test_image.jpg' # hoặc 'test_video.mp4'

# 3. Chạy nhận diện
results = model.predict(source=SOURCE_PATH, conf=0.4, save=True)

print(f"Đã xong! Kết quả lưu tại thư mục runs/detect/predict")
```

```python
import glob
import os

# Tìm ảnh mới nhất trong thư mục predict
latest_prediction = max(glob.glob('runs/detect/predict*/' + SOURCE_PATH), key=os.path.getmtime)
display(Image(filename=latest_prediction))
```
